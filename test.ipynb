```python
import itk

fixed_image_path = "Data/case6_gre1.nrrd"
moving_image_path = "Data/case6_gre2.nrrd"

PixelType = itk.F  # Utilisation de itk.F pour float
Dimension = 3
ImageType = itk.Image[PixelType, Dimension]

fixed_image = itk.imread(fixed_image_path, ImageType)
moving_image = itk.imread(moving_image_path, ImageType)

print(f"Fixed Image Type: {type(fixed_image)}")
print(f"Moving Image Type: {type(moving_image)}")
```
